```python
import pandas as pd
import json
import glob
import os
```

```python
all_json = glob.glob(r"C:\Users\AKBARS\Desktop\Data\*.json")
```

```python
all_json
```

```python
# Список для хранения данных
rows = []

# Проход по всем файлам в папке
for file_path in all_json:
    # Загрузка данных из JSON-файла
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    
    # Предположим, что данные в JSON-файле находятся в ключе 'refs'
    refs = data.get('refs', [])  
    
    # Парсинг данных
    for ref in refs:
        # Проверка, что ref не None и является списком
        if ref is not None and isinstance(ref, list):
            content = ref[0]  
            date_info = ref[1]  
            stats = ref[2]          
            date_str = f"{date_info.get('day')} {date_info.get('month')} {date_info.get('time')}"
            
            # Добавление данных в список
            rows.append({
                'content': content,
                'date': date_str.strip(),  
                'rate': stats.get('rate', None),
                'views': stats.get('views', None),
                'name_file': file_path.split("\\")[-1] })
```

```python
# Создание DataFrame
df = pd.DataFrame(rows)
```

```python
# Сохранение DataFrame в CSV
df.to_csv('json-convert.csv', index=False)
```
